```python
df["Last_Purchase_Date"] = pd.to_datetime(df["Last_Purchase_Date"])
```

```python
latest_Date =df["Last_Purchase_Date"].max()
```

```python
df["days_Since_Last_Purchase"] = (
    latest_Date - df["Last_Purchase_Date"]
).dt.days
```

```python
df[["Customer_ID", "Last_Purchase_Date", "days_Since_Last_Purchase"]].head()
```

```python
df["Risk"] = np.where(
    (df["days_Since_Last_Purchase"] > 180) &
    (df["Total_Orders"] < 5) &
    (df["Lifetime_Value"] < 10000),
    "High Risk",
    "Low Risk"
)
```

```python
df["Lifetime_Value"] = (
    df["Lifetime_Value"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["Lifetime_Value"] = pd.to_numeric(df["Lifetime_Value"])
```

```python
df["Risk"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Risk"].value_counts().plot(kind="bar")

plt.title("Customer Risk Distribution")
plt.xlabel("Risk")
plt.ylabel("Number of Customers")

plt.show()
```

```python
features = df[[
    "Age",
    "Annual_Income",
    "Total_Orders",
    "Lifetime_Value",
    "days_Since_Last_Purchase"
]]

target = df["Risk"]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

target = le.fit_transform(target)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
df["Annual_Income"] = (
    df["Annual_Income"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

df["Annual_Income"] = pd.to_numeric(df["Annual_Income"], errors="coerce")
```

```python
df["Annual_Income"] = pd.to_numeric(df["Annual_Income"], errors="coerce")
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, prediction)

print("Accuracy :", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, prediction)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, prediction))
```

```python
importance = pd.DataFrame({
    "Feature": features.columns,
    "Importance": model.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

print(importance)
```
